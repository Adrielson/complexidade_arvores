```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# Carregar os dados do CSV
data = pd.read_csv('resultados.csv')

# Criar a pasta 'plots' se no existir
os.makedirs('plots', exist_ok=True)

# Funo para plotar grficos separados para cada caso e operao
def plot_by_case_and_operation(case, operation, label):
    plt.figure(figsize=(12, 8))
    
    algorithms = ['BST', 'AVL']
    colors = ['r', 'b']  # Adicionar mais cores se necessrio

    for idx, algorithm in enumerate(algorithms):
        df_algo_case = data[(data['Algoritmo'] == algorithm) & (data['Caso'] == case)]
        plt.plot(df_algo_case['Tamanho'], df_algo_case[operation], marker='o', 
                 label=f'{algorithm}', color=colors[idx])
    
    plt.title(f'Operao de {label} para {case} Caso')
    plt.xlabel('Tamanho da entrada')
    plt.ylabel('Tempo (ns)')
    plt.legend()
    plt.grid(True)
    plt.xticks([2000, 4000, 6000, 8000, 10000], ['2000', '4000', '6000', '8000', '10000'])
    plt.savefig(f'plots/{case.lower().replace(" ", "_")}_{label.lower()}.png')
    plt.show()
    plt.close()

# Gerar grficos para cada caso e operao
# plot_by_case_and_operation('Pior - Crescente', 'Tempo Insercao (ns)', 'Insero')
# plot_by_case_and_operation('Medio', 'Tempo Insercao (ns)', 'Insero')
# plot_by_case_and_operation('Pior - Crescente', 'Tempo Busca (ns)', 'Busca')
# plot_by_case_and_operation('Medio', 'Tempo Busca (ns)', 'Busca')
# plot_by_case_and_operation('Pior - Crescente', 'Tempo Remocao (ns)', 'Remoo')
# plot_by_case_and_operation('Medio', 'Tempo Remocao (ns)', 'Remoo')

plot_by_case_and_operation('Pior', 'Tempo Insercao (ns)', 'Insero')
plot_by_case_and_operation('Medio', 'Tempo Insercao (ns)', 'Insero')
plot_by_case_and_operation('Pior', 'Tempo Busca (ns)', 'Busca')
plot_by_case_and_operation('Medio', 'Tempo Busca (ns)', 'Busca')
plot_by_case_and_operation('Pior', 'Tempo Remocao (ns)', 'Remoo')
plot_by_case_and_operation('Medio', 'Tempo Remocao (ns)', 'Remoo')
```
