```python
import os
import pandas as pd
import matplotlib.pyplot as plt

# Carregar os dados do CSV
data = pd.read_csv('comparacao_arvores.csv')

# Criar a pasta 'plots' se não existir
os.makedirs('plots', exist_ok=True)

# Função para plotar gráficos separados para cada caso e operação
def plot_by_case_and_operation(case, operation, label):
    plt.figure(figsize=(12, 8))
    
    algorithms = ['BST', 'AVL']
    colors = ['b', 'g']  # Adicionar mais cores se necessário

    for idx, algorithm in enumerate(algorithms):
        df_algo_case = data[(data['Algoritmo'] == algorithm) & (data['Caso'] == case)]
        plt.plot(df_algo_case['Tamanho'], df_algo_case[operation], marker='o', 
                 label=f'{algorithm}', color=colors[idx])
    
    plt.title(f'Desempenho dos Algoritmos - {case.capitalize()} Caso - {label}')
    plt.xlabel('Tamanho da entrada')
    plt.ylabel('Tempo (ns)')
    plt.legend()
    plt.grid(True)
    plt.xticks([2000, 4000, 6000, 8000, 10000], ['2000', '4000', '6000', '8000', '10000'])
    plt.savefig(f'plots/desempenho_algoritmos_{case.lower()}_{label.lower()}.png')
    plt.show()
    plt.close()

# Gerar gráficos para cada caso e operação
plot_by_case_and_operation('pior', 'Tempo Insercao (ns)', 'Inserção')
plot_by_case_and_operation('medio', 'Tempo Insercao (ns)', 'Inserção')
plot_by_case_and_operation('pior', 'Tempo Busca (ns)', 'Busca')
plot_by_case_and_operation('medio', 'Tempo Busca (ns)', 'Busca')
plot_by_case_and_operation('pior', 'Tempo Delecao (ns)', 'Deleção')
plot_by_case_and_operation('medio', 'Tempo Delecao (ns)', 'Deleção')
```
